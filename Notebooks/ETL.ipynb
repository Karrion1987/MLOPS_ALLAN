**ETL CSVdata_reviews_final**

```python
#Crea la lista vacía para almacenar el csv
filas = []

# Abre el archivo en modo lectura
with open('/content/drive/MyDrive/MLOPS - ALLAN/2- CSV/CSV_ETL/data_reviews_final.csv', 'r') as archivo_csv:
    # Crea un objeto lector CSV
    lector_csv = csv.reader(archivo_csv)
    encabezados = next(lector_csv)
    # Itera sobre las filas del archivo CSV
    for fila in lector_csv:
        filas.append(fila)

# Transpone la lista de filas para convertirla en columnas
df_reviews = pd.DataFrame(filas, columns=encabezados)

# Muestra el DataFrame
df_reviews.head()
```

```python
df_reviews.info()
```

```python
# Reemplaza los valores en blanco del dataframe por NA
df_reviews.replace('', pd.NA, inplace=True)
```

```python
# Elimina las filas que tiene todos los valores nulos
df_reviews=df_reviews.dropna(how='all')
df_reviews
```

```python
# Elimina las filas que tienen valores nulos en la columna item_id
df_reviews = df_reviews.dropna(subset=['item_id'])
```

```python
# Transformacion de los tipos de datos
df_reviews['item_id']=df_reviews['item_id'].astype(int)
df_reviews['recommend']=df_reviews['recommend'].astype(bool)
df_reviews
```

```python
df_reviews.info()
```

```python
# Muestra el porcentajes de datos nulos por columnas
porcentajes_nulos = (df_reviews.isnull().mean() * 100).round(2)
porcentajes_nulos
```

```python
# Elimina las columnas funny y last_edited
data_reviews = df_reviews.drop(['funny', 'last_edited'], axis=1)
```

```python
data_reviews
```

```python
# Muestra el porcentajes de datos nulos por columnas
porcentajes_nulos = (data_reviews.isnull().mean() * 100).round(2)
porcentajes_nulos
```

```python
#Crea la lista vacía para almacenar el csv
filas = []

# Abre el archivo en modo lectura
with open('/content/drive/MyDrive/MLOPS - ALLAN/2- CSV/CSV_ETL/data_games_final.csv', 'r') as archivo_csv:
    # Crea un objeto lector CSV
    lector_csv = csv.reader(archivo_csv)
    encabezados = next(lector_csv)
    # Itera sobre las filas del archivo CSV
    for fila in lector_csv:
        filas.append(fila)

# Transpone la lista de filas para convertirla en columnas
df_games = pd.DataFrame(filas, columns=encabezados)

# Muestra el DataFrame
df_games.tail(100)
```

```python
valores_distintos = df_games['price'].unique()

# Muestra los valores distintos
print(valores_distintos)
```

```python
df_games.replace(['Free To Play', 'Free to Play', 'Free', 'Play for Free!', 'Free to Use' ], 0, inplace=True)
```

```python
# crea una mascara para eliminar los valores que no son numeros y los nulos
mask = df_games['price'].notna() & df_games['price'].astype(str).str.contains(r'^-?\d+\.?\d*$')
df_games = df_games[mask]
df_games
```

```python
df_games['price']= df_games['price'].astype(float)
```
